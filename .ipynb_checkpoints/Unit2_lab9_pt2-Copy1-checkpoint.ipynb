### Write a function that checks if customer borrowed more or less films in the month of June as compared to May.
#### Hint: For this part, you can create a join between the two dataframes created before, using the merge function available for pandas dataframes.

```python
# checking both dataframes
rentals_may.head()
```

```python
rentals_june.head()
```

```python
# df merging both May and June rentals on customer_id
may_june = rentals_may.merge(rentals_june, how="outer", on='customer_id', suffixes = ('_m', '_j'))
may_june
```

```python
# new df to get the needed columns like count of rentals per customer_id
rent = pd.DataFrame(may_june.groupby('customer_id')[pd.Series(['rental_id_m', 'rental_id_j'])].count())
rent
```

```python
# copying df... just in case!
rent_diff = rent.copy()
rent_diff.reset_index(inplace=True)
```

```python
rent_diff
```

```python
rent_diff['diff'] = pd.Series(rent_diff.rental_id_j - rent_diff.rental_id_m)
rent_diff
```

```python
rent_diff.isna().any()
```

```python
def rent_change(col):
    '''This function will check the difference between the customer rentals between the months of May and June.
    Results will be as follows in the column rent:
    up = June rentals are higher than May rentals
    down = June rentals are lower than May rentals
    equal = No change / Same rentals on both months
    '''
    change = []
    for i in col:
        if i > 0:
            change.append('up')
        elif i < 0:
            change.append('down')
        else:
            change.append('equal')
    return pd.Series(change)
rent_diff['change'] = rent_change(rent_diff['diff'])
```

```python
# checking df
rent_diff.sample(10)
```

```python
rent_diff.change.value_counts()
```
